```python
%pip install pandas
import pandas as pd

#Se extraen los datos en bruto
df = pd.read_excel('paquete_1_datos_bruto.xlsx')
#df = pd.read_csv('paquete_1_datos_bruto_1.csv')

#Se crea un dataframe con los valores de longitud (L), parámetro (variable) e intensidad (value)
df2 = pd.melt(df, id_vars = ['L'])
#Se renombran las columnas:
df2.columns = ['medida_longitud', 'medida_parametro', 'medida_intensidad']
#, 'experimento_operador', 'experimento_instrumento', 'experimento_fecha', 'experimento_fecha_comp', 'idexperimento']

#Se eliminan los registros sin medidas de intensidad
df2 = df2.dropna(subset=["medida_intensidad"])

#Se eliminan los registros con un guión en la medida de intensidad
df2.drop(df2[(df2["medida_intensidad"] == '-')].index, inplace=True)


df2.to_csv('paquete_1_datos_modificado.csv')
df2
```
